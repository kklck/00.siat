```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

''' 0.  figure()  +  +  '''
# plt.figure()
# plt.plot(np.random.randn(100))
# plt.plot(np.random.randn(200).cumsum())
# plt.show

""" 1.  figure(=?)   subplot    """
fig = plt.figure(figsize=(16,8)) # 
ax1 = fig.add_subplot(2,1,1) #  , 1 2 1,     
ax2 = fig.add_subplot(2,1,2) # 2
ax1.plot(np.random.randn(100))
ax2.plot(np.random.randn(200).cumsum())
```

```python
""" 2. figure   (pyplot) subplot()    """
fig = plt.figure(figsize=(16,8)) # 
ax1 = plt.subplot(2,1,1) #  , 1 2 1
ax2 = plt.subplot(2,1,2) # 2
ax1.plot(np.random.randn(100))
ax2.plot(np.random.randn(200).cumsum())
```

```python
""" 3.  ax   """
fig, ax = plt.subplots(2,1)
ax[0].plot(np.random.randn(100))
ax[1].plot(np.random.randn(200).cumsum())
```
