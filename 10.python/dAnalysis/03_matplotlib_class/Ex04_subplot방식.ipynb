```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

''' 0.  figure(그림영역)을 만들고 + 그리고 + 보여주기 '''
# plt.figure()
# plt.plot(np.random.randn(100))
# plt.plot(np.random.randn(200).cumsum())
# plt.show

""" 1.  figure(그림틀=공간?)을 만들고 거기에 subplot을 추가 하는 방식 """
fig = plt.figure(figsize=(16,8)) # 공간
ax1 = fig.add_subplot(2,1,1) # 부분 영역, 1행 2열 1번째, 번째는 위치가 아닌 영역의 순서
ax2 = fig.add_subplot(2,1,2) # 2번째
ax1.plot(np.random.randn(100))
ax2.plot(np.random.randn(200).cumsum())
```

```python
""" 2. figure로 공간을 만들고 파이플롯(pyplot)의 subplot()으로 각각의 패널 지정 """
fig = plt.figure(figsize=(16,8)) # 공간
ax1 = plt.subplot(2,1,1) # 부분 영역, 1행 2열 1번째
ax2 = plt.subplot(2,1,2) # 2번째
ax1.plot(np.random.randn(100))
ax2.plot(np.random.randn(200).cumsum())
```

```python
""" 3. 간단하게 ax 객체를 이용하기 """
fig, ax = plt.subplots(2,1)
ax[0].plot(np.random.randn(100))
ax[1].plot(np.random.randn(200).cumsum())
```
