```python
%matplotlib inline
    
import matplotlib.pyplot as plt
import numpy as np

""" 0. 플롯 스타일 """
# plt.style.use("classic")
plt.style.use("seaborn-whitegrid")

x = np.linspace(0,10,100) # 0~10 사이 100으로 나눈
plt.plot(np.sin(x))
```

```python
""" 1. 색상지정 """
plt.plot(np.sin(x-0), color="blue");
plt.plot(np.sin(x-1), color="g"); # green
plt.plot(np.sin(x-2), color="#FFDD55"); # 16진수 RGB
plt.plot(np.sin(x-3), color="chartreuse"); # html 색상명
plt.plot(np.sin(x-4), color="0.75"); # 명도
plt.plot(np.sin(x-5), color=(1.0,0.5,0.3));
```

```python
""" 2. 선스타일 """
plt.plot(np.sin(x-0), linestyle="-"); # solid
plt.plot(np.sin(x-1), linestyle="--"); # dashed
plt.plot(np.sin(x-2), linestyle="-."); # dashed + dotted
plt.plot(np.sin(x-3), linestyle=":"); # dotted
plt.plot(np.sin(x-6), linestyle="solid");
```

## r(red) g(green) b(blue) c(cyan) m(magenta) y(yellow) k(black) 

```python
""" 3. 선과 스타일을 간결하게 (대략 읽는 수준으로 ) """
plt.plot(np.sin(x-0),"-g") # solid / green
plt.plot(np.sin(x-1),"--c") # dashed / cyan
plt.plot(np.sin(x-2),"-.k") # dashdot / black
plt.plot(np.sin(x-3),":r") # dotted / red
```

```python
''' 4. 범례 '''
plt.plot(np.sin(x-0),"-g", label="first") # solid / green
plt.plot(np.sin(x-1),"--c", label="second") # dashed / cyan
plt.plot(np.sin(x-2),"-.k", label="third") # dashdot / black
plt.plot(np.sin(x-3),":r", label="fourth") # dotted / red


plt.legend()
```
